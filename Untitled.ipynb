```python
# Import necessary Qiskit libraries
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import numpy as np
```

```python
# Function to implement the Quantum Fourier Transform (QFT)
def qft(n):
    qc = QuantumCircuit(n)
    
    # Apply the QFT circuit
    for qubit in range(n):
        qc.h(qubit)  # Apply Hadamard gate to each qubit
        for j in range(qubit + 1, n):
            qc.cp(np.pi / 2**(j - qubit), qubit, j)  # Controlled Phase gate
    
    # Swap qubits to reverse the order
    for i in range(n // 2):
        qc.swap(i, n - i - 1)
    
    return qc
```

```python

# Create a 5-qubit QFT circuit
n = 5
qc = qft(n)
```

```python
# Draw the circuit
qc.draw('mpl')  # 'mpl' is for matplotlib visualization
```
